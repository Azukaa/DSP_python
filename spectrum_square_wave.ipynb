```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal, fft, arange
from scipy.fftpack import fft, fftshift
%matplotlib inline
```

```python
f = 10 #Frequency of square wave
time = 2 #Length of time
Fs = 1000 #Sampling rate
Ts = 1/float(Fs) # Time interval between samples
tt = np.arange(0,2,Ts) #Create the time vector
sqr = signal.square(2*np.pi*f*tt) #Create the square wave
print(len(sqr))
plt.plot(tt,sqr) #plot the square wave
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
```

```python
#Compute the FFT spectrum
X = fft(sqr)
plt.plot(abs(X))
```
